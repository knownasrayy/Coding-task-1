```python
mean_curah_hujan = df['Curah Hujan (mm)'].mean()
mean_penjualan_A = df['Penjualan A (pcs)'].mean()
mean_penjualan_B = df['Penjualan B (pcs)'].mean()

print("Mean Curah Hujan:", mean_curah_hujan)
print("Mean Penjualan A:", mean_penjualan_A)
print("Mean Penjualan B:", mean_penjualan_B)
```

```python
median_curah_hujan = df['Curah Hujan (mm)'].median()
median_penjualan_A = df['Penjualan A (pcs)'].median()
median_penjualan_B = df['Penjualan B (pcs)'].median()

print("Median Curah Hujan:", median_curah_hujan)
print("Median Penjualan A:", median_penjualan_A)
print("Median Penjualan B:", median_penjualan_B)
```

```python
mode_curah_hujan = df['Curah Hujan (mm)'].mode()[0]
mode_penjualan_A = df['Penjualan A (pcs)'].mode()[0]
mode_penjualan_B = df['Penjualan B (pcs)'].mode()[0]

print("Modus Curah Hujan:", mode_curah_hujan)
print("Modus Penjualan A:", mode_penjualan_A)
print("Modus Penjualan B:", mode_penjualan_B)
```

```python
range_curah_hujan = df['Curah Hujan (mm)'].max() - df['Curah Hujan (mm)'].min()
range_penjualan_A = df['Penjualan A (pcs)'].max() - df['Penjualan A (pcs)'].min()
range_penjualan_B = df['Penjualan B (pcs)'].max() - df['Penjualan B (pcs)'].min()

print("Range Curah Hujan:", range_curah_hujan)
print("Range Penjualan A:", range_penjualan_A)
print("Range Penjualan B:", range_penjualan_B)
```

```python
Q1_curah_hujan = df['Curah Hujan (mm)'].quantile(0.25)
Q3_curah_hujan = df['Curah Hujan (mm)'].quantile(0.75)
IQR_curah_hujan = Q3_curah_hujan - Q1_curah_hujan

Q1_penjualan_A = df['Penjualan A (pcs)'].quantile(0.25)
Q3_penjualan_A = df['Penjualan A (pcs)'].quantile(0.75)
IQR_penjualan_A = Q3_penjualan_A - Q1_penjualan_A

Q1_penjualan_B = df['Penjualan B (pcs)'].quantile(0.25)
Q3_penjualan_B = df['Penjualan B (pcs)'].quantile(0.75)
IQR_penjualan_B = Q3_penjualan_B - Q1_penjualan_B

print("IQR Curah Hujan:", IQR_curah_hujan)
print("IQR Penjualan A:", IQR_penjualan_A)
print("IQR Penjualan B:", IQR_penjualan_B)
```

```python
variance_curah_hujan = df['Curah Hujan (mm)'].var()
variance_penjualan_A = df['Penjualan A (pcs)'].var()
variance_penjualan_B = df['Penjualan B (pcs)'].var()

print("Variance Curah Hujan:", variance_curah_hujan)
print("Variance Penjualan A:", variance_penjualan_A)
print("Variance Penjualan B:", variance_penjualan_B)
```

```python
std_curah_hujan = df['Curah Hujan (mm)'].std()
std_penjualan_A = df['Penjualan A (pcs)'].std()
std_penjualan_B = df['Penjualan B (pcs)'].std()

print("Standard Deviation Curah Hujan:", std_curah_hujan)
print("Standard Deviation Penjualan A:", std_penjualan_A)
print("Standard Deviation Penjualan B:", std_penjualan_B)
```

```python
skewness_curah_hujan = df['Curah Hujan (mm)'].skew()
skewness_penjualan_A = df['Penjualan A (pcs)'].skew()
skewness_penjualan_B = df['Penjualan B (pcs)'].skew()

print("Skewness Curah Hujan:", skewness_curah_hujan)
print("Skewness Penjualan A:", skewness_penjualan_A)
print("Skewness Penjualan B:", skewness_penjualan_B)
```

```python
kurtosis_curah_hujan = df['Curah Hujan (mm)'].kurtosis()
kurtosis_penjualan_A = df['Penjualan A (pcs)'].kurtosis()
kurtosis_penjualan_B = df['Penjualan B (pcs)'].kurtosis()

print("Kurtosis Curah Hujan:", kurtosis_curah_hujan)
print("Kurtosis Penjualan A:", kurtosis_penjualan_A)
print("Kurtosis Penjualan B:", kurtosis_penjualan_B)
```
